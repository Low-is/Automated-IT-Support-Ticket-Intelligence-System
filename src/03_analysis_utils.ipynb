```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


def get_freq_by_category(df, category_col, feature_col):

    freq = df.groupby([category_col, feature_col]).size().reset_index(name="count")

    freq["percent"] = freq.groupby(category_col)["count"].transform(
        lambda x: x / x.sum() * 100
    )

    return freq # return pandas data frame

# Usage
# freq = get_freq_by_category(processed_df, "priority", "type")
# get_freq_by_category(processed_df, "priority", "queue")
# get_freq_by_category(processed_df, "queue", "type")



def plot_category_distribution(freq_df, category_col, feature_col, color="royalblue"):

    plt.figure(figsize=(10,6))

    ax = sns.barplot(
        data=freq_df,
        x=feature_col,
        y="percent",
        hue=category_col,
        dodge=False,
        edgecolor="black",
        palette=[color]
    )

    # ---------------------------
    # Titles and labels
    # ---------------------------
    ax.set_title(f"{feature_col.capitalize()} Distribution by {category_col.capitalize()}",
                 fontsize=14,
                 fontweight="bold")

    ax.set_ylabel("Percentage (%)", fontsize=12)
    ax.set_xlabel(feature_col.capitalize(), fontsize=12)

    plt.xticks(rotation=45, ha="center")

    ax.legend(title=category_col.capitalize())

    # ---------------------------
    # Add value labels
    # ---------------------------
    for p in ax.patches:
        height = p.get_height()

        if height > 0:
            ax.annotate(
                f"{height:.1f}%",
                (p.get_x() + p.get_width()/2, height),
                ha="center",
                va="bottom",
                fontsize=10,
                fontweight="bold"
            )

    plt.tight_layout()
    plt.show()

# Usage
# plot_category_distribution(freq, "priority", "type")
# With custom color
# plot_category_distribution(freq, "priority", "type", color="darkorange")
```
